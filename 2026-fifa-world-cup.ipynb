```python
import seaborn as sns, matplotlib.pyplot as plt
```

```python
# sns.boxplot(data=df, x=df['position'], y=df['shots'])
plt.figure(figsize=(12,6))
sns.boxplot(data=df,x=df['position'], y=df['shots'], fill=True, native_scale=True)
plt.title('Shots by Position')
plt.xlabel('Postion')
plt.ylabel('Number of Shots')
plt.show()
```

```python
sns.histplot(data=df,legend =True, kde=True, x = df['pass_accuracy'])
plt.title('Pass Accuracy of Players')
plt.xlabel('Pass Accuracy')
plt.show()
```
